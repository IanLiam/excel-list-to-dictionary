```python
import pandas as pd

# Replace '/path/to/your/excel/file.xlsx' with the actual path to your Excel file
excel_file_path = r'C:\Users\ian.muthengi\Downloads\model.xlsx'

# Replace 'Column_Name' with the actual column name you want to create a dictionary from
column_name = 'MODEL'

# Read the Excel file into a DataFrame
df = pd.read_excel(excel_file_path)

# Extract the specified column and create a list of unique values
column_values = df[column_name].unique().tolist()

# Initialize an empty dictionary
my_dict = {}

# Iterate through the unique values and add them to the dictionary with some modifications
for value in column_values:
    # Skip values with colons or spaces after commas
    if ':' not in value and ' ," ' not in value:
        # Remove leading and trailing whitespaces
        value = value.strip()
        # Split the value by commas and remove leading/trailing whitespaces from each part
        parts = [part.strip() for part in value.split(',')]
        # Create a dictionary entry with the first part as the key and the second part as the value
        my_dict[parts[0]] = parts[1]

# Print the resulting dictionary
print(my_dict)




```
